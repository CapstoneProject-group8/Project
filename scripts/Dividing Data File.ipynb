```python
import os
import pandas as pd
import snscrape.modules.twitter as sntwitter
```

```python
# Setting variables to be used below
maxTweets = 100000

# Creating list to append tweet data to
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('#covid since:2017-01-01 until:2019-10-31').get_items()):
    if i>maxTweets:
        break
    tweets_list2.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.user.location,tweet.retweetCount,tweet.likeCount,tweet.coordinates,tweet.lang,tweet.hashtags])
```

```python
# Creating a dataframe from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2, columns=['Datetime', 'Tweet Id', 'Text', 'Username','Location','Retweet','Like','Place','Lang','Hashtags'])

# Display first 5 entries from dataframe
tweets_df2
```

```python
tweets_df2.to_csv('tweets_immigrant_nov2019_to_may2022.csv', index=False)
```
